## Probability
1. Mutually Exclusive: İki olay aynı anda gerçekleşmiyorsa karşılıklı dışlayandır. Bu durumda toplam olasılık bireysel olasılıkların toplamıdır. Örn: Zarda 2 veya 5 gelmesi.
2. Non-mutually Exclusive: İki olay aynı anda gerçekleşebiliyorsa toplam olasılık bireysel olasılıkların toplamından kesişimin çıkartılmasıdır. Örn: kupa veya resimli bir kart çekmek
3. Independent Events: Eğer iki olay birbirini etkilemiyorsa bağımsızdır. Bu durumda iki olayın birlikte gerçekleşmesi çarpılarak bulunur.
4. Dependent Events: Eğer iki olayın gerçekleşmesi birbirini etkiliyorsa bağımlıdır. Bu durumda ikinci olayın gerçekleşmesi birinci olaya bağlıdır.

|  Kural  | Formül  |  Örnek  |
|---------|---------|---------|
| Mutually Exclusive  | P(A OR B)=P(A)+P(B) | Zar atıp 2 veya 5 gelmesi  |
| Non-mutually Exclusive  | P(A OR B)=P(A)+P(B) - P(A AND B)  | Kart destesinden kupa veya resimli kart çekme |
| Independent Events  | P(A AND B) = P(A) x P(B)  | Para atıp zar atma |
| Dependent Events | P(A AND B) = P(A) x P(B\|A) | Bir desteden iki As çekme(geri koymadan) |

## Probability Mass Function
- Discrete random variables
- Çağrı merkezi örneği aşağıdaki gibidir

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parametre
lam = 4  # λ değeri (beklenen olay sayısı)

# x değerleri (olasılığı hesaplayacağımız değer aralığı)
x = np.arange(0, 15)

# Poisson PMF
pmf = poisson.pmf(x, lam)

# Çizim
plt.bar(x, pmf)
plt.title("Poisson Distribution PMF")
plt.xlabel("x -> Çağrı Sayısı")
plt.ylabel("P(X = x) -> Olasılık")
plt.show()
```

2. Probability Density Function
- Boy uzunluğu örneği aşağıdaki gibidir.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parametreler
mu = 170   # Ortalama boy (cm)
sigma = 10 # Standart sapma

# x değerleri (örneğin 130 cm - 210 cm arası boylar)
x = np.linspace(130, 210, 500)

# Normal dağılım PDF
pdf = norm.pdf(x, mu, sigma)

# Çizim
plt.plot(x, pdf, color="blue", linewidth=2)
plt.title("Normal Distribution of Heights")
plt.xlabel("Boy (cm)")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
```

3. Cumulative Distribution Function
- Zar atışı için örnek aşağıdaki gibidir

```python
import numpy as np
import matplotlib.pyplot as plt

# Zarın olası değerleri
x = np.arange(1, 7)

# Her bir değer için olasılık (adil zar → hepsi 1/6)
p = np.ones(6) / 6  

# CDF hesapla (kümülatif toplam)
cdf = np.cumsum(p)

# Çizim
plt.step(x, cdf, where="post", color="blue")
plt.scatter(x, cdf, color="red")  # noktaları işaretle
plt.title("CDF - Zar Atışı")
plt.xlabel("Zar Değeri")
plt.ylabel("P(X ≤ x)")
plt.ylim(0, 1.05)
plt.grid(True)
plt.show()
```

## Bernoulli Dağılımı
- 0,1 gibi iki farklı sonucu olan olaylar için kullanılır. Başarı p başarısızlık 1-p olarak modellenir.

<div>

- Madeni para örneği için grafik aşağıdaki gibidir

```python
from scipy.stats import bernoulli
# Parametre
p = 0.5

# Olası değerler (0 = başarısızlık, 1 = başarı)
x = [0, 1]

# PMF değerleri
pmf = bernoulli.pmf(x, p)

# Çizim
plt.bar(x, pmf, color="skyblue", edgecolor="black", width=0.4)
plt.title(f"Bernoulli Distribution (p={p})")
plt.xlabel("x")
plt.ylabel("P(X = x)")
plt.xticks([0, 1])
plt.ylim(0, 1)
plt.show()
```

## Binom Dağılımı
- n adet Bernoulli denemesinden oluşur
<div style="text-align: center;">
  <img src="https://www.onlinemathlearning.com/image-files/binomial-distribution-formula.png" width="400"/>    
<div>


```python
from scipy.stats import binom

# Parametreler
n = 10   # Atış sayısı
p = 0.6  # Başarı (isabet) olasılığı

# 0'dan n'e kadar olası başarı sayıları
x = np.arange(0, n+1)

# PMF (olasılıklar)
pmf = binom.pmf(x, n, p)

# Çizim
plt.bar(x, pmf, color="orange", edgecolor="black")
plt.title(f"Basket Atışı - Binomial Distribution (n={n}, p={p})")
plt.xlabel("Başarılı Atış Sayısı")
plt.ylabel("P(X = x)")
plt.xticks(x)
plt.show()
```

> Binom dağılımı, bağımsız denemelerde başarı sayısını modellemek için idealdir

## Log-Normal Dağılım 
Bir değişkenin logaritmasının normal dağıldığı bir sürekli olasılık dağılımıdır.

- Örn: Gelir dağılımı

```python
from scipy.stats import lognorm

# Parametreler
mu = 10       # Ortalama (log-scale)
sigma = 0.5   # Standart sapma (log-scale)

# x değerleri (gelir aralığı)
x = np.linspace(0, 50000, 500)  

# PDF
pdf = lognorm.pdf(x, sigma, scale=np.exp(mu))

# Çizim
plt.plot(x, pdf, color="purple", linewidth=2)
plt.title("Gelir Dağılımı - Log-Normal Distribution")
plt.xlabel("Gelir")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
```

## Merkezi Limit Teoremi
Büyük örneklemlerle çalışıldığında, bağımsız rastgele değişkenlerin toplamı veya ortalamasının yaklaşık olarak normal dağıldığını söyler.

- Başlangıçta ne olursa olsun örneklem büyüklüğü arttıkça dağılımın şekli normal dağılıma yaklaşır.

```python
# Parametreler
n_samples = 1000   # toplam tekrar
sample_size = 30   # her seferinde kaç değer örneklenecek

# Uniform dağılımdan örnekleme (0-1 arası)
means = []
for _ in range(n_samples):
    sample = np.random.uniform(0, 1, sample_size)
    means.append(np.mean(sample))

# Histogram çiz
plt.hist(means, bins=30, density=True, color="skyblue", edgecolor="black")
plt.title(f"Central Limit Theorem - Sample Means (n={sample_size})")
plt.xlabel("Örneklem Ortalaması")
plt.ylabel("Yoğunluk")
plt.show()
```

## Hipotez Testi
Bir örneklemden elde edilen verilerin belirli bir iddiayı destekleyip desteklemediğini belirlemek için kullanılan bir yöntemdir.

1. Null Hipotezi(H0): "Değişiklik yok" veya "Etki yok"
2. Alternatif Hipotez(H1): "Farklılık var" veya "Etki var"

- Bir anlamlılık seviyesi seçilir, genellikle 0.05
- Test istatiği seçilir: Z-testi, t-testi, Ki-kare testi
- p hesaplanır: H0'ın doğru olma olasılığı
- p < anlamlılık seviyesi ise H0 reddedilir

## Z-Testi
- Popülasyon standart sapması bilindiğinde ve örneklem büyüklüğü büyük olduğunda kullanılır.
- Tek kuyruklu veya çift kuyruklu uygulanabilir.

> Tek kuyruklu(One-Tailed): Hipotezin belirli bir yönde değişiklik olup olmadığına bakar. Örn: Yeni makine eskisinden hızlı mı?

> Çift Kuyruklu(Two-Tailed): Hipotezin herhangi bir yönde anlamlı bir farklı olup olmadığına bakar. Örn: Yeni makine eskisinden farklı mı? 

## T-Testi
- Popülasyon standart sapması bilinmediğinde ve örneklem küçük olduğunda kullanılır.
- Tek kuyruk ve çift kuyruk z-testi ile aynıdır.
- Bağımsız iki örneklem testi, iki grubun arasındaki farkları belirlemek için kullanılır.
- Eşleştirilmiş t-testi, aynı grup içinde değişimi ölçmek için kullanılır

## Ki-Kare Testi
- Bir veri setinin beklenen dağılıma uyup uymadığı kontrol edilir.
- Örnek olarak zar atışları incelenir, sonuçların adil olup olmadığı kontrol edilir
